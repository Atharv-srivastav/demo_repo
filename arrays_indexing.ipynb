NUMPY INDEXING AND SLICING OF VECTOR

```python
arr =  np.arange(11,21)
arr
```

```python
arr[5]
```

```python
arr[1:5]
```

```python
arr_slice = arr[1::2]
arr_slice
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
slice = arr[0:2,1:3]
slice
```

```python
arr[3]
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
arr[:,2]
```

BOOLEAN INDEXING

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr %2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
